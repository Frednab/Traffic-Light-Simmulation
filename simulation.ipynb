```python
import pygame

# Initialize pygame
pygame.init()

# Set the window size
size = (800, 600)

# Create the window
screen = pygame.display.set_mode(size)


# Initialize the car positions
car_up_x = 440
car_up_y = 435
truck_l_x = 200
truck_l_y = 370
bus_left_x = 100
bus_left_y = 370
car_r_x = 500
car_r_y = 250
car_d_x = 310
car_d_y = 120

# Load the image
bg_image = pygame.image.load("intersection.png")
traffic_red = pygame.image.load("images/signals/red.png")
traffic_yellow = pygame.image.load("images/signals/yellow.png")
traffic_green = pygame.image.load("images/signals/green.png")
traffic4 = pygame.image.load("images/signals/green.png")

#Car images loaded
car_up   = pygame.image.load("images/Up/caru.png")
truck_l  = pygame.image.load("truck.png")
bus_left = pygame.image.load("bus.png")
car_r    = pygame.image.load("images/left/carr.png")
car_d   = pygame.image.load("images/down/card.png")


# Scale the image to fit the window
bg_image       = pygame.transform.scale(bg_image, size)
traffic_red    = pygame.transform.scale(traffic_red,    (50, 100))
traffic_yellow = pygame.transform.scale(traffic_yellow, (50, 100))
traffic_green  = pygame.transform.scale(traffic_green,   (50, 100))
traffic4       = pygame.transform.scale(traffic4,       (50, 100))


# Set the initial traffic light to red
traffic_light = traffic_red


# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        

    # Draw the image on the window
    screen.blit(bg_image, (0, 0))
    screen.blit(traffic_light, (230, 80))
    screen.blit(traffic_light, (490, 80))
    screen.blit(traffic_light, (230, 445))
    screen.blit(traffic_light, (495, 445))
    screen.blit(car_up, (car_up_x, car_up_y))
    screen.blit(truck_l, (truck_l_x, truck_l_y))
    screen.blit(bus_left, (bus_left_x, bus_left_y))
    screen.blit(car_r, (car_r_x, car_r_y))
    screen.blit(car_d, (car_d_x, car_d_y))
    
    
    
    # Update the display
    pygame.display.flip()
    
    
    # Increment the car positions
    car_up_y -= 15
    truck_l_x += 2
    bus_left_x += 2
    car_r_x -= 20
    car_d_y += 25
    
    # Use a timer to change the traffic light every 5 seconds
    pygame.time.wait(5000)

    # Change the traffic light to yellow
    if traffic_light == traffic_red:
        traffic_light = traffic_yellow
    # Change the traffic light to green
    elif traffic_light == traffic_yellow:
        traffic_light = traffic_green
    # Change the traffic light to red
    elif traffic_light == traffic_green:
        traffic_light = traffic_red



# Exit pygame
pygame.quit()
```
